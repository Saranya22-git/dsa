```python
for i in range(1,5):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(1,5):
    for j in range(i):
        print(i, end=" ")
    print()
```

```python
for i in range(1,5):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
for i in range(4,0,-1):
    for j in range(i):
        print("*", end=" ")
    print()
```
