```python
# Check whether the number is even or not

num=7

if num%2==0:
    print("Even")
else:
    print("Odd")
```

```python
# Check whether the number is positive, negative or zero

num=0

if num>0:
    print("Positive")

elif num<0:
    print("Negative")

else:
    print("Zero")
```

```python
# Nested Condition

age=20
has_id=True

if age>=18:
    if has_id==True:
        print("Allowed")
    else:
        print("No ID")
else:
    print("Under Age")
```

```python
# Loop

for i in range(1,6):
    print(i)
```
