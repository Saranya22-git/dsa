```python
# Check whether the number is even or not

num=7

if num%2==0:
    print("Even")
else:
    print("Odd")
```

```python
# Check whether the number is positive, negative or zero

num=0

if num>0:
    print("Positive")

elif num<0:
    print("Negative")

else:
    print("Zero")
```

```python
# Nested Condition

age=20
has_id=True

if age>=18:
    if has_id==True:
        print("Allowed")
    else:
        print("No ID")
else:
    print("Under Age")
```

```python
# Count even numbers from 1 to 10

count=0

for i in range(1,11):
    if i%2==0:
        count=count+1
print(count)
```

```python
# Sum of even numbers

sum=0

for i in range(1,11):
    if i%2==0:
        sum=sum+i
print(sum)
```

```python
# Largest number(3, 7, 2, 9, 5)

largest=3

numbers=[3,7,2,9,5]
for i in numbers:
    if i>largest:
        largest=i
    
print(largest)
```

```python
# Sum of digits

n=482
total=0

while n>0:
    digit=n%10
    total=total+digit

    n=n//10

print(total)
```

```python
# Reverse a number

n=482
rev=0


while n>0:
    digit=n%10
    rev=rev*10+digit

    n=n//10

print(rev)
```

```python
# Palindrome number

n=121

original=n
rev=0

while n>0:
    digit=n%10
    rev=rev*10+digit
    n=n//10
    
if original==rev:
    print("Palindrome")
else:
    print("Not palindrome")
```
